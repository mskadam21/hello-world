```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python

#### ----- Step 1 (Download data)----
URL_DATASET = r'https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv'
```

```python
df1 = pd.read_csv(URL_DATASET)
# print(df1.head(3))  # Uncomment to see the dataframe
```

```python

#### ----- Step 2 (Select data for India)----
df_india = df1[df1['Country'] == 'India']
print(df_india.tail(3))
```

```python

#### ----- Step 3 (Plot data)----
# Increase size of plot
plt.rcParams["figure.figsize"]=20,20  # Remove if not on Jupyter
# Plot column 'Confirmed'
df_india.plot(kind = 'bar', x = 'Date', y = 'Confirmed', color = 'blue')

ax1 = plt.gca()
df_india.plot(kind = 'bar', x = 'Date', y = 'Deaths', color = 'red', ax = ax1)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from plotly import tools
import plotly.plotly as py
import plotly.figure_factory as ff
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
df = pd.read_csv('/home/mkadam/Desktop/Inn-48/Covid19India.org/state_wise.csv') ##,index_col=0, parse_dates=True)
```

```python
df.head()
```

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1['% of Population'] = df1['Confirmed'] / df1['Population'] *100
```

```python
df1['% of Population']
```

```python
df1.plot(kind = 'bar', x = 'State', y = '% of Population', sort_columns=True)
plt.show()
```

```python
Population_df = df1.reset_index().sort_values(by=('Population'),ascending=False)
```

```python
plt.figure(figsize = (12, 10), facecolor = None)
sns.set_style("darkgrid")
sns.set(font_scale=1.5)
plot1 = sns.barplot(x="State", y="Population", data=Population_df)

plot1.set_xticklabels(Population_df['State'], rotation=90, ha="center")
plot1.set(xlabel='State',ylabel='State Population 2019')
plot1.set_title('Current Population Data')
```
